```python
x = df.drop("purchase_status", axis=1)
y = df["purchase_status"]
```

```python
x.shape, y.shape
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    x, y,
    test_size=0.2,
    random_state= 42,
    stratify= y
)
```

```python
x_test.shape, x_train.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
x_train.info()
```

```python
num_cols = ['age','annual_income','number_of_purchases',
            'time_spent_on_website','discounts_availed']
```

```python
scaler = StandardScaler()
x_train[num_cols] = scaler.fit_transform(x_train[num_cols])
x_test[num_cols] = scaler.transform(x_test[num_cols])
```

```python
x_train.head()
```

## Model Training

### logistic regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
model_log_reg = LogisticRegression(max_iter=1000,
                                   class_weight="balanced")
```

```python
model_log_reg.fit(x_train, y_train)
```

```python
y_pred_lr = model_log_reg.predict(x_test)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred_lr))
print("confusion_matrix:\n",confusion_matrix(y_test, y_pred_lr))
print("classification_report:\n",classification_report(y_test, y_pred_lr, zero_division= 0))
```

### random forest classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model_rf = RandomForestClassifier(
    n_estimators= 300,
    random_state= 42,
    class_weight="balanced"
)
```

```python
model_rf.fit(x_train, y_train)
```

```python
y_pred_rf = model_rf.predict(x_test)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred_rf))
print(confusion_matrix(y_test, y_pred_rf))
print(classification_report(y_test, y_pred_rf,zero_division= 0))
```

### KNN 

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(x_train)
X_test_scaled = scaler.transform(x_test)
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train_scaled, y_train)
```

```python
y_pred_knn = knn.predict(X_test_scaled)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred_knn))
print("\nClassification Report:\n", classification_report(y_test, y_pred_knn))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred_knn))
```

## Model Comparision

```python
model_comparison = pd.DataFrame({
    "Model": ["Logistic Regression", "Random Forest", "KNN"],
    "Accuracy": [
        accuracy_score(y_test, y_pred_lr),
        accuracy_score(y_test, y_pred_rf),
        accuracy_score(y_test, y_pred_knn)
    ]
})
```
